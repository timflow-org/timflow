```python
import matplotlib.pyplot as plt
import numpy as np

import timflow.transient as tft
```

```python
k = [20, 40]  # m/d
z = [1, 0, -10, -12, -20]
Ss = 1e-4  # 1/m
c = [20, 500]  # d
tsandhstar = [(0, 1), (10, 0)]
```

```python
def model(beta=0):
    ml = tft.ModelXsection(naq=2, tmin=0.0001, tmax=100)

    tft.XsectionMaq(
        model=ml,
        x1=-np.inf,  # river extends to infinitiy
        x2=0,
        z=z,
        kaq=k,
        Saq=Ss,
        Sll=Ss,
        beta=beta,
        c=c,
        topboundary="semi",
        tsandhstar=tsandhstar,
        name="river",
    )

    tft.XsectionMaq(
        model=ml,
        x1=0,
        x2=np.inf,  # hinterland extends to infinity
        kaq=k,
        z=z[1:],
        c=c[1:],
        Saq=Ss,
        topboundary="conf",
        name="land",
    )

    ml.solve()
    return ml
```

```python
x = np.linspace(-300, 300, 100)
t = [1 / 1000, 1, 24]
ml = model(beta=0)
h = ml.headalongline(x, 0, t).squeeze()
plt.figure(figsize=(5, 2))
for i in range(len(t)):
    plt.plot(x, h[0, i], "C" + str(i))
    plt.plot(x, h[1, i], "C" + str(i))
plt.xlim(-300, 300)
plt.ylim(0, 1)
plt.grid()
```

```python
t = np.logspace(-4, 2, 100)
ml = model(beta=0)
h = ml.head(0, 0, t)
plt.semilogx(t, h[0], label="lay=0, beta=0")
plt.semilogx(t, h[1], label="lay=1, beta=0")
ml = model(beta=1)
h = ml.head(0, 0, t)
plt.semilogx(t, h[0], label="lay=0, beta=1")
plt.semilogx(t, h[1], label="lay=1, beta=1")
plt.legend()
plt.grid()
```
