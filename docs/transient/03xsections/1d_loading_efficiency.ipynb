# 1D loading efficiency

Test for loading efficiency.

```python
import numpy as np

import timflow.transient as tft
```

```python
z = [1, 0.0, -5.0, -6.0, -15.0]
Saq = [0.2, 1e-4]
c = [10.0, 20]
k = [5.0, 10.0]
delh = 1.0
beta = [0.8, 0.3]
```

```python
ml = tft.ModelMaq(
    kaq=k,
    z=z,
    Saq=Saq,
    c=c,
    beta=beta,
    topboundary="semi",
    phreatictop=True,
    tmin=1e-2,
    tmax=1e2,
)

inhom_left = tft.XsectionMaq(
    ml,
    x1=-np.inf,
    x2=10.0,
    kaq=k,
    z=z,
    Saq=Saq,
    c=c,
    beta=beta,
    topboundary="semi",
    phreatictop=True,
)
```

```python
ml.aq.beta
```
