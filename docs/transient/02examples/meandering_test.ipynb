# Meandering river

```python
import numpy as np

import timflow.transient as tft
```

```python
ml = tft.Model3D(
    kaq=[2, 1, 5, 10, 4],
    z=[10, 8, 6, 4, 2, 0],
    Saq=[0.1, 0.0001, 0.0002, 0.0002, 0.0001],
    phreatictop=True,
    kzoverkh=0.1,
    topboundary="conf",
    tmin=1,
    tmax=10,
    M=1,
)
w = tft.Well(ml, -25, 0, rw=0.3, tsandQ=[(0, 100)], layers=[2, 3])

xy = [(-50, -30), (40, 20)]

#ls1 = tft.RiverString(ml, xy=xy, tsandh="fixed", layers=[0, 1])
ls1 = tft.River(ml, -50, -30, 40, 20, tsandh="fixed", layers=[1])
```
