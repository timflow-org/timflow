### Well string

```python
import matplotlib.pyplot as plt
import numpy as np

import timflow.transient as tft

plt.rcParams["font.size"] = 8.0
plt.rcParams["figure.figsize"] = (8, 3)
```

```python
# T = 500  # transmissivity
# S = 1e-4
# t = np.logspace(-2, 2, 101)
# Q = 1000
# xy=[(-10, 0), (10, 0)]
```

```python
import timflow.transient as tft
ml = tft.ModelMaq(kaq=[10, 20], z=[20, 10, 8, 0], c=[200], Saq=1e-4, tmin=0.01, tmax=100)
w = tft.Well(ml, 0, 0, tsandQ=[(0, 1000)], rw=0.3, layers=[0,1])
ml.solve()
w.headinside(7)
```

```python
import timflow.transient as tft
ml = tft.ModelMaq(kaq=[10, 20], z=[20, 10, 8, 0], c=[200], Saq=1e-4, tmin=0.01, tmax=100)
w = tft.WellString(ml, xy=[(0, 10), (0, -10)], tsandQ=[(0, 1000)], rw=0.3, layers=[0,1])
ml.solve()
w.wlist[0].headinside(7), w.wlist[1].headinside(7)
```
