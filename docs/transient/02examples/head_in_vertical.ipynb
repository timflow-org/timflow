# Head propagation in leaky layers

```python
import matplotlib.pyplot as plt
import numpy as np

import timflow.transient as tft

plt.rcParams["font.size"] = 8.0
```

Consider a three-aquifer system. Aquifer properties are given in Table 1. All aquifers have elastic storage. A well is located at $(x,y)=(0,0)$ and is screened in layer 1. The well starts pumping at time $t=0$ with a discharge $Q=1000$ m$^3$/d. The radius of the well is 0.2 m.

**Table 1 - Aquifer properties for exercise 1**
|Layer          | $k$ (m/d) | $c$ (d) | $S_s$ (m$^{-1}$) | $z_t$ (m) | $z_b$ (m)|
|---------------| ---------:| -------:| -----:| ---------:| --------:|
|Aquifer 0      |      1    |    -    |0.0001 |   25      |        20|
|Leaky layer 1  |      -    |  1000   |0      |   20      |        18|
|Aquifer 1      |     20    |    -    |0.0001 |   18      |        10|
|Leaky layer 2  |      -    |  2000   |0      |   10      |         8|
|Aquifer 2      |      2    |    -    |0.0001 |    8      |         0|


## Confined top

```python
ml = tft.ModelMaq(
    kaq=[5, 10, 20],
    z=[25, 22, 18, 13, 8, 0],
    c=[1000, 2000],
    Saq=[1e-4, 1e-4, 1e-4],
    Sll=[1e-3, 1e-3],
    phreatictop=False,
    tmin=0.1,
    tmax=1000,
)
w1 = tft.Well(ml, xw=0, yw=0, rw=0.2, tsandQ=[(0, 1000)], layers=1)
w2 = tft.Well(ml, xw=20, yw=0, rw=0.2, tsandQ=[(0, 177)], layers=1)
ml.solve()
```

```python
ml.head(10, 10, 1)
```

```python
z = np.linspace(0, 25, 100)
h = ml.headinvertical(10, 10, z, [0.1, 1])
```

```python
z = np.linspace(0, 25, 100)
t = [0.1, 1, 10, 100]
h = ml.headinvertical(10, 10, z, t)
for i in range(len(t)):
    plt.plot(h[:, i], z)
for i in range(ml.aq.naq - 1):
    plt.axhspan(ml.aq.zaqtop[i + 1], ml.aq.zaqbot[i], color=[0.9, 0.9, 0.9])
plt.axhspan(25, 30, color=[0.6, 0.6, 0.6])
plt.ylim(z[0], 30);
```

## Semi-confined top

```python
ml = tft.ModelMaq(
    kaq=[5, 10, 20],
    z=[30, 25, 22, 18, 13, 8, 0],
    c=[500, 1000, 2000],
    Saq=[1e-4, 1e-4, 1e-4],
    Sll=[1e-3, 1e-3, 1e-3],
    topboundary="semi",
    tmin=0.1,
    tmax=1000,
)
w1 = tft.Well(ml, xw=0, yw=0, rw=0.2, tsandQ=[(0, 1000)], layers=1)
w2 = tft.Well(ml, xw=20, yw=0, rw=0.2, tsandQ=[(0, 177)], layers=1)
ml.solve()
```

```python
z = np.linspace(0, 30, 100)
t = [0.1, 1, 10, 100]
h = ml.headinvertical(10, 10, z, t)
for i in range(len(t)):
    plt.plot(h[:, i], z)
for i in range(len(ml.aq.z[::2]) - 1):
    plt.axhspan(ml.aq.z[2 * i + 1], ml.aq.z[2 * i], color=[0.9, 0.9, 0.9])
plt.ylim(z[0], z[-1]);
```

## Leaky layer top

```python
ml = tft.ModelMaq(
    kaq=[5, 10, 20],
    z=[30, 25, 22, 18, 13, 8, 0],
    c=[500, 1000, 2000],
    Saq=[1e-4, 1e-4, 1e-4],
    Sll=[1e-3, 1e-3, 1e-3],
    topboundary="lea",
    tmin=0.1,
    tmax=1000,
)
w1 = tft.Well(ml, xw=0, yw=0, rw=0.2, tsandQ=[(0, 1000)], layers=1)
w2 = tft.Well(ml, xw=20, yw=0, rw=0.2, tsandQ=[(0, 177)], layers=1)
ml.solve()
```

```python
z = np.linspace(0, 30, 100)
t = [0.1, 1, 10, 100]
h = ml.headinvertical(10, 10, z, t)
for i in range(len(t)):
    plt.plot(h[:, i], z)
for i in range(len(ml.aq.z[::2]) - 1):
    plt.axhspan(ml.aq.z[2 * i + 1], ml.aq.z[2 * i], color=[0.9, 0.9, 0.9])
plt.ylim(z[0], z[-1]);
```
